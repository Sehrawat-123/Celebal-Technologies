```python
import pandas as pd
from matplotlib import pyplot as plt

# Insert the CSV file
sample_data = pd.read_csv('AAPL.csv')

sample_data['Date'] = pd.to_datetime(sample_data['Date'])

print(sample_data.head())

print(type(sample_data))

# Plot the Volume over Date
plt.figure(figsize=(12, 6))
plt.plot(sample_data['Date'], sample_data['Volume'], 'o', label='Volume')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Apple Stock Volume over Time')
plt.legend()
plt.show()

# Plot High and Low prices
plt.figure(figsize=(12, 6))
plt.plot(sample_data['Date'], sample_data['High'], label='High')
plt.plot(sample_data['Date'], sample_data['Low'], label='Low')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Apple Stock High and Low Prices over Time')
plt.legend()
plt.show()
```
